```python
def shows_plot_for_town (town):
    df_amk = df[df['town']== town]

    df_amk_3RM = df_amk[df_amk['flat_type']=='3-RM']
    df_amk_3RM_time = df_amk_3RM['quarter']

    df_amk_3RM_time
    df_amk_1RM = df_amk[df_amk['flat_type']=='1-RM']
    df_amk_1RM_time = df_amk_1RM['quarter']

    df_amk_2RM = df_amk[df_amk['flat_type']=='2-RM']
    df_amk_2RM_time = df_amk_2RM['quarter']

    df_amk_4RM = df_amk[df_amk['flat_type']=='4-RM']
    df_amk_4RM_time = df_amk_4RM['quarter']
    
    df_amk_5RM = df_amk[df_amk['flat_type']=='5-RM']
    df_amk_5RM_time = df_amk_5RM['quarter']

    df_amk_EXEC = df_amk[df_amk['flat_type']=='EXEC']
    df_amk_EXEC_time = df_amk_EXEC['quarter']
    
    plt.plot(df_amk_1RM_time,df_amk_1RM['median_rent'],color ='m', alpha = 0.8)
    plt.plot(df_amk_2RM_time,df_amk_2RM['median_rent'],color ='c', alpha = 0.8)
    plt.plot(df_amk_3RM_time,df_amk_3RM['median_rent'], color ='red', alpha = 0.8)
    plt.plot(df_amk_5RM_time,df_amk_4RM['median_rent'],color ='violet', alpha = 0.8)
    plt.plot(df_amk_5RM_time,df_amk_5RM['median_rent'],color ='blue', alpha = 0.8)
    plt.plot(df_amk_EXEC_time,df_amk_EXEC['median_rent'],color ='green', alpha = 0.8)
    plt.title("Rent Prices for " + town)
    plt.xlabel("Quarter")
    plt.ylabel("Median Rent for " + town)
    plt.legend(["AMK 1-RM","AMK 2-RM","AMK 3-RM","AMK 4-RM", "AMK 5-RM",'AMK EXEC'])
    plt.show()
```
